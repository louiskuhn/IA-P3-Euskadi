## Éléments explicatifs sur la différence entre Dense et Sparse Vectors

Ce sont seulement des formats de vecteurs. Les vecteurs denses sont les vecteurs tels qu'on les connaît, par exemple pour un vecteur avec 3 valeurs : `[val1, val2, val3]`.  
Les vecteurs sparse sont utilisés pour mieux gérer la mémoire. En effet, on renseigne seulement les éléments différents de 0 et leurs positions dans le vecteur. En ajoutant la longueur du vecteur, l'objet dispose de toutes les méta-informations pour reconstituer le vecteur. Le vecteur s'écrit alors `nb_elements, [position1, position2, position3], [val1, val2, val3]`.  
Encore une fois, seules les positions et les valeurs différentes de 0 sont indiquées. Dans le cas où on a un gros vecteur avec beaucoup de 0, il y aura un gain important de mémoire à stocker le vecteur sous cette forme.   


__Exemples :__

```python
from pyspark.ml.linalg import SparseVector, DenseVector 
a = SparseVector(4, [1, 3], [3.0, 4.0])
b = DenseVector(a.toArray())

print(a, b)
```

En général, on n'a pas à se soucier de la différence entre ces vecteurs. Dans MLlib par exemple, les fonctions sont capables d'interpréter les deux de manière identique. On a pu constater par exemple en créant notre colonne `features` que celle-ci s'affichait sous forme de vecteur sparse dans le dataframe, mais il n'y a rien de spécial à faire pour qu'ensuite le modèle de régression linéaire interprète correctement les valeurs de ce vecteur comme s'il était dense.
