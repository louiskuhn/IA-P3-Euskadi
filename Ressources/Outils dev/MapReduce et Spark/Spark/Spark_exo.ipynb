#### 2. c) Dans un notebook, chargez le fichier CSV en PySpark dans un dataframe

```
# File location and type
file_location = "/FileStore/tables/transactions.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)
```

```
df = spark.read.csv(file_location, header=first_row_is_header, inferSchema=infer_schema,  sep=delimiter)
```

## 3. Du SQL depuis Databricks
#### 3. a) À l'aide de la méthode df.createOrReplaceTempView, mettez le dataframe dans une table

```
df.createOrReplaceTempView("transac")
```

#### 3. b) Refaites les calculs de la question 2.f) en SQL

```
%sql
SELECT avg(amount), sum(amount), stddev(amount), min(amount), max(amount) FROM transac;
```

```
%sql
SELECT * FROM transac ORDER BY amount DESC LIMIT 5;
```

```
%sql
SELECT country_sender, avg(amount), sum(amount) FROM transac GROUP BY country_sender;
```

```
%sql
SELECT sent, received, sent-received FROM (
  (SELECT account_sender_name, sum(amount) AS sent FROM transac WHERE account_sender_name = "Laurent Alexandre" GROUP BY account_sender_name) AS a
  LEFT JOIN
  (SELECT account_receiver_name, sum(amount) AS received FROM transac WHERE account_receiver_name = "Laurent Alexandre" GROUP BY account_receiver_name) AS b
  ON a.account_sender_name = b.account_receiver_name
);
```

## 4. Sauvegarde de la table/df
#### Créez une table avec les transactions émises depuis la France et l’enregistrez pour pouvoir y accéder ailleurs que dans ce notebook

```
%sql
DROP TABLE IF EXISTS transac_fr;
CREATE TABLE transac_fr AS SELECT * FROM transac WHERE country_sender = "FR";
SELECT * FROM transac_fr;
```

```
df_fr = df.filter(df["country_sender"] == "FR")
display(df_fr)
```

```
# en table SQL
df_fr.write.mode('overwrite').saveAsTable('transac_fr_from_df')
```

```
%sql
SELECT * FROM transac_fr_from_df;
```

```
# au format parquet
df_fr.write.format('delta').save('/FileStore/delta/transac_fr')
```

```
# lire un fichier parquet
display(spark.read.load('/FileStore/delta/transac_fr'))
```
