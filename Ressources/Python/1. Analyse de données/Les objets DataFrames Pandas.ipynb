## Méthode Apply() sur les colonnes d'un DataFrame

- Obtenir le dataframe float_df qui ne contient que les colonnes de type décimales float64
>- dtypes est votre ami
>- vaut-il mieux utiliser le dataframe fandango ou bien fandango_perso (indexés par les films) ?
- Diviser par 2 chaque valeur, élever chaque valeur au carré, calculer l'écart-type de chaque colonne
>- apply et la fonction std de la limbraire numpy sont vos amis
>- stocker les résultats dans 2 variables

```python
# solution avec méthode select_dtypes
float_df = fandango.select_dtypes(include="float64").copy()
```

```python
filtre = fandango.dtypes =="float64"
float_df = fandango[fandango.columns[filtre]].copy()
float_df
```

```python
def div(x):
    return x/2

div(float_df.iloc[:, 0])
float_df.apply(div, axis=0)
```

```python
# définition "à la volée"
float_df.apply(lambda x : x/2)
```

```python
# au carré
float_df.apply(lambda x : x**2)
```

```python
import numpy as np
float_df.apply(np.std, axis=0)
```
