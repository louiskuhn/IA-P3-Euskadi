## Analyse des données

Regarder l'évolution des différentes informations fournies pour chaque joueur et comparer les afin d'en savoir plus sur ce qui peut influencer les salaires.

Pour cela, commencer par construire une fonction __*courbes(data,joueurs)*__ :
- qui prend en argument :
>- *data* : les données que l'on veut afficher (par exemple Salary ou encore FreeThrows/FreeThrowsAttempts pour le taux de réussite aux lancer-francs ou ce qu'on veut)
>- *joueurs* : la liste de joueurs dont on veut afficher les informations
- qui retourne :
>- une figure avec l'ensemble des courbes de chaque joueur sur le même graphique

Cette fonction vous permettra d'avoir une première analyse des données et de leur évolution en affichant les données sous forme de courbes.

### Des pistes pour l'analyse

Vous trouverez quelques pistes possibles pour l'analyse ci-dessous.

- Des problématiques auxquelles répondre :
>- évolution des salaires
>- comparaison des joueurs selon des mesures "dans le jeu" (les paniers, les points, les tentatives,...)
>- évolution et impact des temps de jeu
>- les styles de jeu :
>>- précision des joueurs
>>- plus de paniers à 2 points/3 points
>>- impact des lancer-francs

- Des questions techniques à vous poser :
>- Regarder d'abord les données telles qu'elles sont, sans autre manipulation. Qu'en dire ?
>- Y a-t-il des valeurs aberrantes ? Que sont-elles ? Que veulent-elles dire ? Comment peut-on les gérer ?
>- Regarder ensuite des informations comparables : cela revient à "normaliser" les données. Comment faire ?

```python
import matplotlib.pyplot as plt

def courbe(data, joueurs):
    for joueur in joueurs:
        idx = Players.index(joueur)
        plt.plot(Seasons, data[idx], label=joueur)
    plt.legend()
    plt.show()
```

```python
courbe(Games, ['KobeBryant', 'JoeJohnson', 'DerrickRose'])
```

```python
courbe(FreeThrows, ['KobeBryant', 'JoeJohnson', 'DerrickRose'])
```

```python
courbe(FreeThrows/FreeThrowsAttempts, ['KobeBryant', 'JoeJohnson', 'DerrickRose'])
```
